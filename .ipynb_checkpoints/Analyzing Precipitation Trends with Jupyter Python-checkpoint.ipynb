# Best Fit Distribution

Best fit distribution for precipitation.

```python
# first we trim days without precipitation and the higher 0.2%
precipcut = datos['Ppt'][datos['Ppt']>0][datos['Ppt']<datos['Ppt'].quantile(0.998)]

# we have to calculate mean and standar deviation
promedio = precipcut.mean()
desviacion = precipcut.std()
print (promedio, desviacion)
```

```python
# now we determine the probabilistic distributions
tabulaciones = np.arange(-40,51,0.1)
distnormal = stats.norm.pdf(tabulaciones, 
                            loc=promedio, scale=desviacion)
distlognormal = stats.pearson3.pdf(tabulaciones,skew=1,
                                   loc=promedio, scale=desviacion)
distweibull = stats.dweibull.pdf(tabulaciones,c=1,
                                   loc=promedio, scale=desviacion)
distchi2 = stats.chi2.pdf(tabulaciones,df=2,
                                   loc=promedio, scale=desviacion)
```

```python
# we plot the normalized histogram and the probabilistic distributions
precipcut.hist(bins=20, 
                  normed=True, edgecolor='#4aaaaa', color='#80BCD8')
plot(tabulaciones,distnormal, color='#4B4C4E', linewidth=5, linestyle='--',label='Dist Normal')
plot(tabulaciones,distlognormal, color='#3F83B7', linewidth=5, linestyle='--', label='Dist Lognormal')
plot(tabulaciones,distweibull, color='#7B7C7E', linewidth=5, linestyle='-.', label='Dist Weibull')
plot(tabulaciones,distchi2, color='#3F83B7', linewidth=5, linestyle=':', label='Dis Chi2')
xlim(0,50)
legend(loc='upper right')
figsize(21,14)
```

# Finding sequence of days

Need to find sequence of days having this range, for eg: day 1 has 30 mm of rainfall,  day 2, day 3, day 4 has 2 mm of rainfall. 

```python
datos['Ppt'].count()
```

```python
datos['Value'] = 0
for i in arange(datos['Ppt'].count()-3):
    if datos['Ppt'][i] > 30:
        if datos['Ppt'][i+1] < 2 and datos['Ppt'][i+2] < 2 and datos['Ppt'][i+3] < 2:
            datos['Value'][i]=1
            print "Secuence of days were found at: " + str(datos['Ppt'].index[i])  
            print "Ppt fo day 1: " + str(datos['Ppt'][i]) + ' Ppt for day 2,3,4: ' + \
            str(datos['Ppt'][i+1]) + ' ' + str(datos['Ppt'][i+2])+ ' ' + str(datos['Ppt'][i+3])
```

```python
datos[datos['Value']>0]['Ppt'].ix['1960-01-01':'1989-12-31'].plot.density()
```
